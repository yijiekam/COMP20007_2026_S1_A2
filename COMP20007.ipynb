```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

csv_url = 'https://raw.githubusercontent.com/yijiekam/COMP20007_2026_S1_A2/refs/heads/main/cmsExperiments.csv'
cms_df = pd.read_csv(csv_url)


# processing out the columns we don't need
cms_df = cms_df[1:]
cms_df = cms_df.iloc[:, 11:].reset_index(drop=True)

# making the first row the header
cms_df.columns = cms_df.iloc[0]

# drop first row, not needed
cms_df = cms_df[1:11]

# convert values to int so can be graphed properly

y1 = cms_df['avg. acc.'].astype(float)
y2 = cms_df['avg fp. '].astype(float)

x = np.arange(len(y1)) + 1

plt.plot(x, y1, color='blue')
plt.title('Average accuracy over different numbers of tables')
plt.xlabel('Number of tables')
plt.ylabel('Average accuracy')
plt.xticks(x);

# other plot for false positives
#plt.plot(x, y2, color='red')
#plt.title('Average false positives over different numbers of tables')
#plt.xlabel('Number of tables')
#plt.ylabel('Average false positives')
#plt.xticks(x);
```
