```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

csv_url = 'https://raw.githubusercontent.com/yijiekam/COMP20007_2026_S1_A2/refs/heads/main/cmsExp_num_tbls_raw.csv'
cms_numtbls_df = pd.read_csv(csv_url)


# convert values to int so can be graphed properly

y1 = cms_numtbls_df['avg_acc'].astype(float)
y2 = cms_numtbls_df['avg_fp '].astype(float)

x = np.arange(len(y1)) + 1

#plt.plot(x, y1, color='blue')
#plt.title('Average accuracy over different numbers of tables')
#plt.xlabel('Number of tables')
#plt.ylabel('Average accuracy')
#plt.xticks(x);

# other plot for false positives
plt.plot(x, y2, color='red')
plt.title('Average false positives over different numbers of tables')
plt.xlabel('Number of tables')
plt.ylabel('Average false positives')
plt.xticks(x);

```
